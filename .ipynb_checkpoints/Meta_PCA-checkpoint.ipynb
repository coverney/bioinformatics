```python
# import needed libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import plotly.plotly as py
from plotly.graph_objs import *
import plotly.tools as tls
import plotly.graph_objs as go
tls.set_credentials_file(username='enter info', api_key='enter info')
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__)
init_notebook_mode(connected=True)
```

```python
#Create a DataFrame containing the data you wish to analyze.
#In this case it is the GH metagenomics data
df = pd.read_csv(
    filepath_or_buffer='meta_data.csv', 
    header=None, 
    sep=',')
df
```

```python
#Show another way to get a Numpy representation of a Dataframe
D = df.ix[:,0:21].values
print(D)
```

```python
#Center the transpose of the data matrix.
#In other words, center the rows of the data matrix
B = StandardScaler(copy=True, with_mean=True, with_std=False).fit_transform(D.T)
#Get the centered data matrix
B = B.T
```

```python
print(B)
```

```python
#Divide each entry in the data matrix by the standard deviation of the samples in that row.
x = np.matrix(df.ix[:,0:21].values.reshape((17, 21)))
B = B/x.std(1, ddof=1)
B.round(2)
```

```python
#Next, we must create the covariance matrix.
C = (B).dot(B.T) / (B.shape[1]-1)
print('Covariance matrix: \n%s' %C.round(3))
```

```python
#Find the eigenvectors and eigenvalues of the covariance matrix.
eig_vals, eig_vecs = np.linalg.eig(C)

print('Eigenvectors \n%s' %eig_vecs.round(2))
print('\nEigenvalues \n%s' %eig_vals.round(2))
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0].round(2))
    print(i[1].round(2))
```

```python
#Format a plot to show how much of the total variance each principal component explains.
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

trace1 = Bar(
        x=['PC %s' %i for i in range(1,18)],
        y=var_exp,
        showlegend=False)

trace2 = Scatter(
        x=['PC %s' %i for i in range(1,18)], 
        y=cum_var_exp,
        showlegend=False)

data = Data([trace1, trace2])

layout=Layout(
        yaxis=YAxis(title='Explained Variance(%)'),
        title='Explained Variance by Principal Components',
        font=dict(size=16)
        )

fig = Figure(data=data, layout=layout)
py.iplot(fig, filename='ex_var')
```

```python
#Creates a matrix containing the two dominant eigenvectors as column vectors.
V = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1)))

print('Matrix V:\n', V.round(2))
```

```python
#Creates a matrix containing the three dominant eigenvectors as column vectors.
V2 = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1), 
                eig_pairs[2][1].reshape(17,1)))

print('Matrix V2:\n', V2.round(2))
```

```python
#Dimensionally reduce the data matrix into 2d
B2 = B.T.dot(V)
print(B2)
```

```python
#Dimensionally reduce the data matrix into 3d
B3 = B.T.dot(V2)
print(B3)
```

```python
#Plot the three dominant principal components from PCA on a 3d axis
traces = []

trace1 = go.Scatter3d(
    x=B3[:8,0],
    y=B3[:8,1],
    z=B3[:8,2],
    mode='markers',
    text=['FCW18b', 'FCF4', 'FCF6TC3', 'FCP16c', 'FCS17a', 'FCC6', 'FCX16C', 'FCR15b'],
    hoverinfo = 'text',
    name='Freshwater Samples',
    marker=Marker(
            size=12,
            line=Line(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5),
            opacity=0.8))

traces.append(trace1)

trace2 = go.Scatter3d(
    x=B3[8:15,0],
    y=B3[8:15,1],
    z=B3[8:15,2],
    mode='markers',
    text=['SCB2a_77', 'SCB14c', 'SCR11c', 'SCQ2a', 'SCD17c', 'SCQ17a', 'SCH14a'],
    hoverinfo = 'text',
    name='Marine Samples',
    marker=Marker(
            size=12,
            line=Line(
                color='green',
                width=0.5),
            opacity=0.8))

traces.append(trace2)

trace3 = go.Scatter3d(
    x=B3[15:,0],
    y=B3[15:,1],
    z=B3[15:,2],
    mode='markers',
    text = ['45591', '66722', '148651', '145604', '66718', '55268'],
    hoverinfo = 'text',
    name='Environmental References',
    marker=Marker(
            size=12,
            line=Line(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5),
            opacity=0.8))

traces.append(trace3)

data = Data(traces)

layout = Layout(showlegend=True,
                scene=Scene(xaxis=XAxis(title='PC1'),
                yaxis=YAxis(title='PC2'),zaxis=ZAxis(title='PC3')),
                title='3D PCA of GH Data',
                margin=dict(t=50, b=50, r=0, l=0, autoexpand=False)
               )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3d_PCA_GH')
```
