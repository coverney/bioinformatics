```python
# libraries for PCoA
from scipy.spatial import distance
# skbio's PCoA library used for verification
from skbio.stats.ordination import pcoa
from skbio.diversity import beta_diversity
```

```python
#Create a DataFrame containing the data you wish to analyze.
#In this case it is the GH metagenomics data
df = pd.read_csv(
    filepath_or_buffer='meta_data.csv', 
    header=None, 
    sep=',')
df
```

```python
#Show another way to get a Numpy representation of a Dataframe
D = df.ix[:,0:21].values
print(D)
```

```python
# Compute the Bray-curtis distance between all combinations of sample pairs and place values into a distance matrix
# The diagonal values will all be 0 because a variable is no distance away from itself
B = []
for row in D:
    B_row = []
    for row2 in D:
        B_row.append(distance.braycurtis(row, row2))
    B.append(B_row)
B = np.matrix(B)
B.round(2).reshape(17, 17)
```

```python
# eigenvalues from skbio's PCoA function, used for verification
pcoa(B).eigvals
```

```python
# First transformation of the distance matrix
A = -0.5*np.square(B)
A.round(2)
```

```python
# computing the mean row vector and column vector
A_col = A.mean(0)
A_row = A.mean(1)
print(A_col)
print(A_row)
```

```python
# compute the mean of the entire matrix A
A_mean = A.mean()
A_mean
```

```python
# Second transformation of the distance matrix (standardization)
E = A - A_col - A_row + A_mean
E
```

```python
#Find the eigenvectors and eigenvalues of E.
eig_vals, eig_vecs = np.linalg.eig(E)

print('Eigenvectors \n%s' %eig_vecs.round(2))
print('\nEigenvalues \n%s' %eig_vals.round(2))
```

```python
# Scales the eigenvactors such that their lengths are equal to their 
# corresponding eigenvalues
eig_vecs_scaled = []
for i in range(len(eig_vals)):
    eig_vec = eig_vecs[:,i]
    eig_vecs_scaled.append(eig_vec*eig_vals[i])
eig_vecs_scaled
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs_scaled[i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0].round(2))
    print(i[1].round(2))
```

```python
#Creates a matrix containing the two dominant eigenvectors as column vectors.
V = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1),))
#                       eig_pairs[2][1].reshape(17,1)))

print('Matrix V:\n', V.round(2))
```

```python
#Creates a matrix containing the three dominant eigenvectors as column vectors.
V2 = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1), 
                eig_pairs[2][1].reshape(17,1)))

print('Matrix V2:\n', V2.round(2))
```

```python
#Dimensionally reduce the data matrix into 2d
D2 = D.T.dot(V)
print(D2)
```

```python
#Dimensionally reduce the data matrix into 3d
D3 = D.T.dot(V2)
print(D3)
```
